# Fill missing values in the type column.

```python
data["Typ"].value_counts()
```

```python
data["Typ"].isna().value_counts()
```

```python
data.rename(columns={"Typ":"Type"}, inplace=True)
data["Type"].replace("TVShow","TV Show",inplace=True)
data["Type"].replace("TV Sh","TV Show",inplace=True)
```

```python
data
```

```python
def add_missing_values_in_type_col(row):
    if pd.isna(row["Type"]):
        if "min" in row["duration"]:
            return "Movie"
        else:
            return "TV Show"
    return row["Type"]
```

```python
data["Type"]=data.apply(add_missing_values_in_type_col,axis=1)
```

```python
data["Type"].value_counts()
```

```python
data["Type"].isna().sum()
```

```python
data
```

```python
data["director"].value_counts()
```

```python
data[["director","cast","country"]]=data[["director","cast","country"]].fillna("unknown")
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dropna(subset="title",inplace=True)
```

```python
data.isna().sum()
```

```python
data
```

```python
data["date_added"]=pd.to_datetime(data["date_added"],format="%d-%b-%y",errors="coerce")
```

# Visualization

```python
plt.pie(data["Type"].value_counts(),labels=data["Type"].value_counts().index , autopct='%1.1f%%')
plt.title("Movies vs. TV Shows by percentage")
```

```python
plt.barh(data['country'].value_counts().head(10).index,data['country'].value_counts().head(10))
plt.title("Top 10 countries produce the most content on Netflix")
plt.xlabel("number of contributions")
plt.ylabel("country")
```

```python
plt.bar(data["rating"].value_counts().head(10).index,data["rating"].value_counts().head(10))
plt.title("Top 10 most common ratings")
plt.xlabel("rating")
plt.ylabel("counts")
```

```python
plt.barh(data[(data["Type"]=="Movie") & (data["director"] !="unknown")]["director"].value_counts().head(10).index,data[(data["Type"]=="Movie") & (data["director"] !="unknown")]["director"].value_counts().head(10))
plt.title("Top 10 directors who have the most movies")
plt.ylabel("Director")
plt.xlabel("no_contributions")
```

```python
data[data["Type"]=='TV Show']["duration"].value_counts()
```

```python
plt.barh(data[data["Type"]=='TV Show']["duration"].value_counts().index,data[data["Type"]=='TV Show']["duration"].value_counts())
plt.title("The durations of TV Shows")
plt.ylabel("no_Seasons")
plt.xlabel("no_contributions")
```

# Bonus

```python
def actor1(col):
    col=col.split(",")
    return col[0].strip()

data["actor1"]=data["cast"].transform(actor1)
```

```python
def actor2(col):
    col=col.split(",")
    if len(col)==1:
        return "unknown"
    return col[1].strip()

data["actor2"]=data["cast"].transform(actor2)
```

```python
plt.barh(data[data['actor1']!= "unknown"]['actor1'].value_counts().head(10).index,data[data['actor1']!= "unknown"]['actor1'].value_counts().head(10))
plt.title("Top 10 actors appear in Movies/TV Shows")
plt.ylabel("actor")
plt.xlabel("no_contributions")
```
